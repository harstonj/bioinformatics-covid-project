```python
# parse a fasta file
def parse_fasta_file(input_file):
    parsed_seqs = {}
    curr_id = None
    curr_seq = []

    for line in input_file:
        line = line.strip()

        if line.startswith('>'):
            if curr_id is not None:
                parsed_seqs[curr_id] = ''.join(curr_seq)

            curr_id = line[1:]
            curr_seq = []
            continue

        curr_seq.append(line)

    # add final sequence to dict
    parsed_seqs[curr_id] = ''.join(curr_seq)
    return parsed_seqs

#------------------------------------------------------------------------------
# parse a fasta file and replace ids
def parse_fasta_file2(input_file, keys):
    parsed_seqs = {}
    curr_id = None
    curr_seq = []
    i = 0

    for line in input_file:
        line = line.strip()

        if line.startswith('>'):
            if curr_id is not None:
                parsed_seqs[curr_id] = ''.join(curr_seq)

            curr_id = keys[i]
            curr_seq = []
            i+=1
            continue

        curr_seq.append(line)

    # add final sequence to dict
    parsed_seqs[curr_id] = ''.join(curr_seq)
    return parsed_seqs

#------------------------------------------------------------------------------
# print a fasta file
def print_fasta_file(file_name):
#     infile = open(file_name)
    for line in infile:
        print(line.strip())

#------------------------------------------------------------------------------
# write a fasta file
def write_fasta_file(output_path, sequences, line_chars=80):
    output_file = open(output_path, 'w')

    for seq_id, seq in sequences.items():
        id_line = '>' + seq_id + "\n"
        output_file.write(id_line)

        if len(seq) < line_chars:
            sequence_line = seq + "\n"
            output_file.write(sequence_line)
            continue

        seq_lines = [seq[i:i+line_chars] for i in range(0,len(seq),line_chars)]
        for seq_line in seq_lines:
            output_file.write(seq_line + "\n")

    # close file when done
    output_file.close()
```
